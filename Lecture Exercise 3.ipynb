# Plotting Kinematic

# Kinematic equation $x = x_0 + v_0t+\frac{1}{2}at^2$

```python
# Initial conditions
x0 = 0.0 # m  Initial postion
v0 = 0.0 # m/s Initial velocity
a = 2.0 # m/s^2 acceleration (constant)

t, h, n = 0.0, 0.1, 20 # initial time, step size, number of steps
ta, xa = [], []  # time and position list

for i in range(n):
    x = x0 + v0*t + 0.5*a*t**2
    ta.append(t)
    xa.append(x)
    t = t + h
    
# plot reseult
plt.figure(figsize=(4,4))
plt.plot(ta,xa, '-o', color = "blue", linewidth = 2)
plt.xlabel("Time (in seconds)")
plt.ylabel("Distance (in meters)")

#plt.savefig("kinematics.png")
#plt.show()
```

# Kinematic equation Free fall $y = y_0 + v_0t-\frac{1}{2}gt^2$

```python
# Initial conditions
y0 = 10.0 # m  Initial postion
v0 = 0.0 # m/s Initial velocity
g = 9.8 # m/s^2 acceleration (constant)

t, h, n = 0.0, 0.1, 20 # initial time, step size, number of steps
ta, ya = [], []  # time and position list

for i in range(n):
    y = y0 + v0*t - 0.5*g*t**2
    ta.append(t)
    ya.append(y)
    t = t + h
    
# plot reseult
plt.figure(figsize=(4,4))
plt.plot(ta,ya, '-o', color = "blue", linewidth = 2)
plt.xlabel("Time (in seconds)")
plt.ylabel("Distance (in meters)")

#plt.savefig("kinematics.png")
#plt.show()
```

# Gravitational force

Satellite launched into circular orbit around the Earth, it orbits once every $T$ seconds. The altitude $h =\big(\frac{GMT^2}{4\pi^2}\big)^{\frac{1}{3}}-R$


### Derivation: Gravitation
* Use the following relations to derive the above relation

Gravitational force $F=\frac{Gm_1m_2}{r^2}$ and centripetal force $F=\frac{mv^2}{r}$ and noting that the distance ( is circumference) $C=2\pi r$ the velocity can be calculated $v=\frac{2\pi r}{T}$

```python
def altitude(seconds): # defining a function
    G = 6.67e-11 #Graviational Constant
    M = 6e24 #Mass of the Earth
    R = 6371e3
    T =float(seconds)
    
    h = ((G*M*T**2)/(4*np.pi**2))**(1/3)-R
    return h
print(altitude(input("write the period of the satellite's orbit in seconds:")))
```

# Monte Carlo Integration


```python
from numpy.random import RandomState

def f(x):#function to be integrated
    return x**2

#Create ans instance of Random State
t = RandomState()

# Random x and y arrays
n = int(input("Number of Monte Carlo Trials: "))
x = t.rand(n)
y = t.rand(n)

sum = 0
for i in range(n):
    if (y[i]<f(x[i])):
        sum = sum +1 # sum +=1
        
print('{0:d} Monte Carlo Trials'.format(n))
print('Monte Carlo Answer :{0:10.7f} '.format(sum/n))
print('Exact Answer:{0:10.7f} '.format(1./3.))
```

# Random Walk 

```python
# Code for 2D random walk.
import numpy
import pylab
import random

# defining the number of steps
n = 1000 #100000

#creating two array for containing x and y coordinate
#of size equals to the number of size and filled up with 0's
x = numpy.zeros(n)
y = numpy.zeros(n)

# filling the coordinates with random variables
for i in range(1, n):
    val = random.randint(1, 4)
    if val == 1:
        x[i] = x[i - 1] + 1
        y[i] = y[i - 1]
    elif val == 2:
        x[i] = x[i - 1] - 1
        y[i] = y[i - 1]
    elif val == 3:
        x[i] = x[i - 1]
        y[i] = y[i - 1] + 1
    else:
        x[i] = x[i - 1]
        y[i] = y[i - 1] - 1


# plotting :
pylab.title("Random Walk ($n = " + str(n) + "$ steps)")
pylab.plot(x, y)
pylab.savefig("rand_walk"+str(n)+".png",bbox_inches="tight",dpi=600)
pylab.show()
```
